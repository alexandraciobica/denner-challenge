```python
import json
import requests
import yaml

url = "https://ckan.opendata.swiss/api/3/action/package_show?id=bevolkerungdichte-nach-unterschiedlichen-raumlichen-auspragungen-seit-2009"

resp = requests.get(url)
resp.content[:500]
```

```python
resp.status_code
```

```python
resp.encoding
```

```python
json_populden = resp.json()
populden_url = json_populden["result"]["resources"][0]["url"]
```

```python
import pandas as pd

zurich_populden = pd.read_csv(populden_url)
```

```python
zurich_populden
```

```python
zurich_populden_ytd = zurich_populden[
    zurich_populden["RaumKategorie"] == "Stadtquartier"
]
zurich_populden_ytd = zurich_populden_ytd.loc[
    zurich_populden_ytd["StichtagDatJahr"]
    == zurich_populden_ytd["StichtagDatJahr"].max()
]
zurich_populden_ytd = zurich_populden_ytd.dropna()

zurich_populden_ytd.to_csv("zurich_populden_ytd.csv")
zurich_populden_ytd.head()

# Space category (technically: SpaceCategory)	Space category (string). Entire city, urban district, urban district
# Sorting variable for the room (technically: RaumSort)	Sorting variable for space (integer). 0 for the entire city, for city districts: district number times 10 (example: 80 for district 8), district number
# Text variable for the room (technically: RaumLang)	Text variable for space (integer). Name of the city quarters and urban districts
# Year (technically: key date date year)	Year (integer).
# Economic resident population (technically: AnzBestWe)	Number of people, economic resident population (integer).
# Total area (technically: AreaT)	Total area in ha (integer). T = Total
# Land area without forest (technically: AreaL)	Total area without water and forest in ha (integer). L = land area without forest
# Settlement area (technically: FlaecheS)	Settlement area in ha (integer). S = Settlement area (ie building footprint, house perimeter, gardens, parking spaces, water basins)
# Population density, reference: entire area (technically: densityT)	Population density in people per ha (integer), reference: total area
# Population density, reference: land area without forest (technically: densityL)	Population density in people per ha (integer), reference: land area without forest
# Population density, reference: settlement area (technically: densityS)	Population density in people per ha (integer), reference: settlement area

# https://data.stadt-zuerich.ch/dataset/bev_bestand_jahr_bevoelkerungsdichten_flaechen_od5802
```
