```python
import json
import requests
import yaml

url = "https://ckan.opendata.swiss/api/3/action/package_show?id=bevolkerung-nach-zivilstand-geschlecht-alter-und-stadtquartier-seit-1993"

resp = requests.get(url)
resp.content[:500]
```

```python
json_popul = resp.json()
popul_url = json_popul["result"]["resources"][0]["url"]
```

```python
import pandas as pd

zurich_popul = pd.read_csv(popul_url)
```

```python
zurich_popul
```

```python
zurich_popul_ytd = zurich_popul.loc[
    zurich_popul["StichtagDatJahr"]
    == zurich_popul["StichtagDatJahr"].max()
]
zurich_popul_ytd = zurich_popul_ytd.dropna()
zurich_popul_ytd.head(10)

# Event year (technically: key date date year)	Year (integer).
# Age (Sort) (technically: AlterVSort)	Age (integer).
# Age (Cd) (technically: AlterVCd)	Age (integer, code).
# 5-year age group (Sort) (technically: AlterV05Sort)	Sorting key for age group (Numeric).
# 5-year age group (code) (technical: AlterV05Cd)	Age Group Code (Numeric).
# 5-year age group (technically: AlterV05Kurz)	5 year age group (e.g. 0-4, 5-10, etc.)
# 10-year age group (Sort) (technically: AlterV10Sort)	Sorting key for age group (Numeric).
# 10-year age group (code) (technically: AlterV10Cd)	Age Group Code (Numeric).
# 10-year age group (technically: AlterV10Kurz)	10 year age group (e.g. 0-9, 10-19, etc.)
# 20-year age group (Sort) (technically: AlterV20Sort)	Sorting key for age group (Numeric).
# 20-year age group (code) (technically: AlterV20Cd)	Age Group Code (Numeric).
# 20-year age group (technically: AlterV20Kurz)	20-year age group (e.g. 0-19, 20-29, etc.)
# Gender (Code) (technically: SexCd)	Gender (string).
# Gender (long) (technically: SexLang)	Gender (string).
# City district (code) (technically: KreisCd)	City circle (integer, code).
# City circle (long) (technically: KreisLang)	Urban district (string).
# City quarter (sort) (technically: QuarSort)	City quarter (Integer).
# City quarter (code) (technical: QuarCd)	City quarter (integer, code).
# City quarter (long) (technically: QuarLang)	City quarter (String).
# Civil status (grouped) (Sort) (technically: Ziv2Sort)	Civil status, sorting key of the group (integer).
# Civil status (grouped) (code) (technical: Ziv2Cd)	Civil status, group code (integer).
# Civil status (grouped) (Lang) (technically: Ziv2Lang)	Civil status grouped. (Values: 1 - Single, 2 - Married, 3 - Divorced, 4 - Widowed, 5 - In a registered partnership, 6 - In a dissolved partnership, 9 - Unknown).
# Economic population (technically: AnzBestWe)	Economic Population (Integer).

# https://data.stadt-zuerich.ch/dataset/bev_bestand_jahr_quartier_zivilstand_ag_geschl_od3310
```

```python
zurich_popul_ytd_sum = (
    zurich_popul_ytd.groupby(["QuarLang", 
                              "SexLang", 
                              "Ziv2Lang"
                             ])
    .agg("sum")
    .reset_index()
)

civil_dict = {
    'Ledig': 'Single', 
    'Verheiratet': 'Married', 
    'Geschieden': 'Divorced', 
    'Unbekannt': 'Unknown',
    'In eingetragener Partnerschaft': 'In registered partnership', 
    'Aufgelöste Partnerschaft': 'Dissolved partnership',
    'Verwitwet': 'Widowed'
}

gender_dict = {"männlich": "Male", "weiblich": "Female"}

# zurich_popul_ytd_sum = zurich_popul_ytd_sum.replace(
#     {"Ziv2Lang": civil_dict}
# )

zurich_popul_ytd_sum = zurich_popul_ytd_sum.replace({"SexLang": gender_dict})
zurich_popul_ytd_sum = zurich_popul_ytd_sum.replace({"Ziv2Lang": civil_dict})

zurich_popul_ytd_sum.drop(['StichtagDatJahr', 'AlterVSort', 'AlterVCd',
       'AlterV05Sort', 'AlterV05Cd', 'AlterV10Sort', 'AlterV10Cd',
       'AlterV20Sort', 'AlterV20Cd', 'SexCd', 'KreisCd', 'QuarSort', 'QuarCd',
       'Ziv2Sort', 'Ziv2Cd'], inplace=True, axis=1, errors='ignore')

zurich_popul_ytd_sum
```

```python
zurich_popul_ytd_sum = (
    zurich_popul_ytd_sum.groupby(["QuarLang", 
                              "SexLang"
                             ])
    .agg("sum")
    .reset_index()
)

zurich_popul_ytd_sum
```
