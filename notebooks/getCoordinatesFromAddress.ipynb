```python
import pandas as pd
import gmaps
import math
import requests
```

```python
# add your key here to be able to use the API
# don't forget to delete it before you commit / push
gmaps_key = ""
```

```python
gmaps.configure(api_key=gmaps_key)
```

```python
# get the latitude and longitude based on an address
def get_lati_longi(api_key, address):
    url = 'https://maps.googleapis.com/maps/api/geocode/json'
    params = {
        "address": address,
        "key": gmaps_key
    } 
    response = requests.get(url, params=params)

    if response.status_code == 200:
        data = response.json()
        if data["status"] == "OK":
            location = data["results"][0]["geometry"]["location"]
            lat = location["lat"]
            lng = location["lng"]
            return lat, lng
        else:
            print(f"Error: {data['error_message']}")
            return 0, 0
    else:
        print("Failed to make the request.")
        return 0, 0
```

# Denner stores

```python
# read the CSV(s) with the store locations
denner_branches = pd.read_csv("../data/denner_branches/denner_branches.csv", index_col=0)
denner_branches
```

```python
# adapt the address format for the get_lati_longi function
denner_addresses = denner_branches['street_number'] + ' ' + denner_branches['post_municipality']
denner_addresses
```

```python
# list with latitudes
lats_denner = []
# list with longitudes
long_denner = []

for address in denner_addresses:
    latitude, longitude = get_lati_longi(gmaps_key, address)
    lats_denner.append(latitude)
    long_denner.append(longitude)
```

```python
# add the latitude and longitude to the dataframe
denner_branches['lat'] = lats_denner
denner_branches['long'] = long_denner

# denner_branches
```

```python
# getting the post codes
post_codes=denner_branches['post_municipality'].apply(lambda x:x[0:4])
denner_branches['address'] = denner_addresses
denner_branches['post_codes'] = post_codes
denner_branches['name'] = 'Denner'
# post_codes
```
